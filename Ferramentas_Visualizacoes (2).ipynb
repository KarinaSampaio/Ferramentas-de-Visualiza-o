```python
import sqlalchemy as sa
import pandas as pd
import dask.dataframe as dd
from urllib.parse import quote_plus

# Configuração da conexão com o banco de dados
server = 'ultraflowdbserver.database.windows.net'
database = 'ultraflowdb'
username = 'usuario_ultraflow'
password = '@database.password123!'
driver = 'ODBC Driver 17 for SQL Server'

# String de conexão usando SQLAlchemy com codificação URL
connection_string = f'mssql+pyodbc://{username}:{quote_plus(password)}@{server}/{database}?driver={quote_plus(driver)}'

# Criando a engine SQLAlchemy
engine = sa.create_engine(connection_string)

# Tentando estabelecer uma conexão e ler dados
try:
    # Executando uma consulta SQL com pandas
    query = 'SELECT * FROM suprimentos.fpedidoscompras'
    df_pandas = pd.read_sql(query, engine)

    # Convertendo pandas DataFrame para dask DataFrame
    df_dask = dd.from_pandas(df_pandas, npartitions=3)

    # Exibindo as primeiras linhas do dataframe
    print(df_dask.head())

    # Realizando operações com Dask
    resultado = df_dask.groupby('FORNECEDOR')[['QTDE', 'VAL_MERCADORIA']].mean().compute()
    print(resultado)
except Exception as e:
    print(f"Erro ao conectar: {e}")


```

## Exemplo de Gráfico 1: Distribuição do Valor da Mercadoria

```python
import sqlalchemy as sa
import pandas as pd
import dask.dataframe as dd
from urllib.parse import quote_plus
import matplotlib.pyplot as plt

# Configuração da conexão com o banco de dados
server = 'ultraflowdbserver.database.windows.net'
database = 'ultraflowdb'
username = 'usuario_ultraflow'
password = '@database.password123!'
driver = 'ODBC Driver 17 for SQL Server'

# String de conexão usando SQLAlchemy com codificação URL
connection_string = f'mssql+pyodbc://{username}:{quote_plus(password)}@{server}/{database}?driver={quote_plus(driver)}'

# Criando a engine SQLAlchemy
engine = sa.create_engine(connection_string)

# Tentando estabelecer uma conexão e ler dados
try:
    # Executando uma consulta SQL com pandas
    query = 'SELECT * FROM suprimentos.fpedidoscompras'
    df_pandas = pd.read_sql(query, engine)

    # Convertendo pandas DataFrame para dask DataFrame
    df_dask = dd.from_pandas(df_pandas, npartitions=3)

    # Exibindo as primeiras linhas do dataframe
    print(df_dask.head())
except Exception as e:
    print(f"Erro ao conectar: {e}")
```

## Exemplo de Gráfico 2: Média do Valor da Mercadoria por Fornecedor

```python
# Agrupando os dados por 'REGIONAL_CC' e calculando a média do 'VAL_MERCADORIA'
media_valor_mercadoria = df_dask.groupby('REGIONAL_CC')['VAL_MERCADORIA'].mean().compute()

# Plotando o gráfico de barras
plt.figure(figsize=(14, 7))
media_valor_mercadoria.plot(kind='bar', color='skyblue')
plt.xlabel('Regional')
plt.ylabel('Média do Valor da Mercadoria')
plt.title('Média do Valor da Mercadoria por Regional')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## Exemplo 3: Quantidade Entregue por Regional

```python
# Agrupando os dados por 'REGIONAL_CC' e somando 'QTDE_ENTREGUE'
quantidade_entregue_regional = df_dask.groupby('REGIONAL_CC')['QTDE_ENTREGUE'].sum().compute()

# Plotando o gráfico de barras
plt.figure(figsize=(14, 7))
quantidade_entregue_regional.plot(kind='bar', color='salmon')
plt.xlabel('Regional')
plt.ylabel('Quantidade Entregue')
plt.title('Quantidade Entregue por Regional')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## Exemplo de Gráfico 4: Quantidade de Pedidos por Situação

```python
import sqlalchemy as sa
import pandas as pd
import dask.dataframe as dd
from urllib.parse import quote_plus
import matplotlib.pyplot as plt

# Configuração da conexão com o banco de dados
server = 'ultraflowdbserver.database.windows.net'
database = 'ultraflowdb'
username = 'usuario_ultraflow'
password = '@database.password123!'
driver = 'ODBC Driver 17 for SQL Server'

# String de conexão usando SQLAlchemy com codificação URL
connection_string = f'mssql+pyodbc://{username}:{quote_plus(password)}@{server}/{database}?driver={quote_plus(driver)}'

# Criando a engine SQLAlchemy
engine = sa.create_engine(connection_string)

try:
    # Executando uma consulta SQL com pandas
    query = 'SELECT * FROM suprimentos.fpedidoscompras'
    df_pandas = pd.read_sql(query, engine)
    
    # Verificando se os dados foram lidos corretamente
    print("Primeiras linhas do DataFrame:")
    print(df_pandas.head())

    # Verificando se a coluna SITUACAO_DO_PC existe
    if 'SITUACAO_DO_PC' not in df_pandas.columns:
        raise ValueError("A coluna 'SITUACAO_DO_PC' não existe no DataFrame.")

    # Verificando os valores únicos na coluna SITUACAO_DO_PC
    print("Valores únicos na coluna 'SITUACAO_DO_PC':")
    print(df_pandas['SITUACAO_DO_PC'].unique())

    # Convertendo pandas DataFrame para dask DataFrame
    df_dask = dd.from_pandas(df_pandas, npartitions=3)

    # Calculando a quantidade de pedidos por situação do PC
    quantidade_pedidos_situacao = df_dask['SITUACAO_DO_PC'].value_counts().compute()

    # Verificando os valores calculados
    print("Quantidade de pedidos por situação do PC:")
    print(quantidade_pedidos_situacao)

    # Plotando a quantidade de pedidos por situação do PC
    plt.figure(figsize=(10, 6))
    quantidade_pedidos_situacao.plot(kind='pie', autopct='%1.1f%%')
    plt.title('Quantidade de Pedidos por Situação do PC')
    plt.ylabel('')
    plt.show()

except Exception as e:
    print(f"Erro ao conectar: {e}")








```
